**Understanding the Elbow Curve**

_WCSS:_ This metric represents the sum of the squared distances between each data point and its assigned cluster center (centroid) in a K-Means clustering solution. Lower WCSS generally indicates a better clustering fit, as data points are closer to their respective centroids.
_Elbow Point:_ The elbow curve typically depicts a decrease in WCSS as the number of clusters increases. However, this decrease in WCSS becomes less significant after a certain point, forming an "elbow" shape in the curve. The number of clusters corresponding to this elbow is often considered the optimal value for K.

Here the elbow seems to be around k=3 or k=4. This suggests that 3 or 4 clusters might be a suitable choice for segmenting the diamond prices in your dataset.

X-axis: Values of K (number of clusters)

Y-axis: Within-Cluster Sum of Squared Distances (inertia)

There's a significant decrease in inertia as the number of clusters increases from 1 to 3 or 4. This indicates that these initial cluster separations are meaningful.
The curve starts to flatten out around 3 or 4 clusters. Adding more clusters beyond this point may not provide a significant benefit in terms of reducing inertia, and might lead to overfitting the data.

**Decision Tree:**

The basic idea behind decision trees is to split the data into subsets based on the features that best separate the target variable.

Here's how a decision tree works:

A. Splitting: The decision tree algorithm starts by selecting the best feature from the dataset to split the data into two or more subsets. This process is repeated recursively for each subset until one of the stopping criteria is met.

B. Stopping Criteria: There are several stopping criteria for splitting, including:

1. No further splits are possible (pure leaf nodes): All data points in a node belong to the same class (for classification) or have similar values (for regression).
2. Maximum depth of the tree is reached: The depth of the tree refers to the length of the longest path from the root node to a leaf node.
3. Minimum number of samples in a node: If the number of samples in a node falls below a certain threshold, further splitting is halted.
4. Maximum number of leaf nodes: Limiting the number of leaf nodes helps prevent overfitting.

C. Predictions: Once the tree is built, new data points are classified or predicted by traversing the tree from the root node down to a leaf node. At each internal node, a decision is made based on the value of a feature, and the algorithm moves to the corresponding child node. This process continues until a leaf node is reached, and the majority class in that leaf node is assigned as the predicted class (for classification) or the average target value in that leaf node is predicted (for regression).


*Decision trees have several advantages:*

Easy to understand and interpret: Decision trees can be visualized, making them easy to interpret and explain to non-technical stakeholders.
Can handle both numerical and categorical data: Decision trees can handle a mixture of numerical and categorical features without requiring extensive preprocessing.
Robust to outliers and missing values: Decision trees are robust to outliers and missing values in the data.

Splitting Criteria:

The decision tree makes decisions based on conditions (if-else statements). Each line below the root node represents a decision based on a specific feature and its value.
For example, "| |--- color <= 3.50" represents a decision where if the "color" feature is less than or equal to 3.50, it follows the path to the left child node, otherwise, it follows the path to the right child node.
Leaf Nodes:

The lines ending with "|--- class: x" represent the leaf nodes of the decision tree. These are the terminal nodes where the decision tree makes a final prediction.
For example, "| | | |--- class: 1" means that if the conditions above it are met (clarity <= 2.50, color <= 3.50), the decision tree predicts class 1 for the input data.

**Recursive Structure:**

The decision tree is built recursively, with each node representing a decision based on a feature and its value.

The tree continues to split until certain stopping criteria are met, such as reaching a maximum depth or having no further improvement in impurity measures (e.g., Gini impurity or entropy) for splitting.

```python
# 2.1.1. Determine Value of 'K' in K-Means using Elbow Curve & KMeans-Inertia
# ---------------------------------------------------------------------------
'''
KMeans-Inertia : Sum of Squared Distances of Samples to their closest Cluster Center (Centroid), Weighted by the Sample Weights (if provided)
'''
wcssd = [] # Within-Cluster-Sum-Squared-Distance
nr_clus = range(1,20) # Number of Clusters
for k in nr_clus:
    kmeans = kmclus(n_clusters=k, init='random', random_state=111)
    kmeans.fit(diamond)
    wcssd.append(kmeans.inertia_)
plt.plot(nr_clus, wcssd, marker='x')
plt.xlabel('Values of K')
plt.ylabel('Within Cluster Sum Squared Distance')
plt.title('Elbow Curve for Optimal K')
plt.show()
```
